```python
# Find unique values in the 'Age' column
unique_ages = data['Age'].unique()

# Print the unique values
print("Unique values in the 'Age' column:")
print(unique_ages)
```

```python
# Create a contingency table for Age and Anxiety
contingency_table = pd.crosstab(data['Age'], data['Anxiety'])

# Print the contingency table
print("Contingency Table:")
print(contingency_table)

# Perform the Chi-Square test
chi2_statistic, p_value, _, _ = chi2_contingency(contingency_table)

# Define null and alternative hypotheses
null_hypothesis = "H0: There is no significant relationship between age and anxiety issues."
alternative_hypothesis = "H1: There is a significant relationship between age and anxiety issues."

# Print the results
print("\nChi-square statistic:", chi2_statistic)
print("p-value:", p_value)

# Compare p-value with significance level
alpha = 0.05
if p_value < alpha:
    print(f"\nReject null hypothesis. {alternative_hypothesis}")
else:
    print(f"\nFail to reject null hypothesis. {null_hypothesis}")

# Calculate degrees of freedom
degrees_freedom = (contingency_table.shape[0] - 1) * (contingency_table.shape[1] - 1)
print(f"\nDegrees of Freedom: {degrees_freedom}")

# Critical value from the chi-square distribution table
from scipy.stats import chi2
critical_value = chi2.ppf(1 - alpha, degrees_freedom)
print(f"Critical Value: {critical_value}")

# Compare chi-square statistic with critical value
if chi2_statistic > critical_value:
    print(f"Reject null hypothesis. {alternative_hypothesis}")
else:
    print(f"Fail to reject null hypothesis. {null_hypothesis}")
```

```python
# Function to perform Chi-Square test
def perform_chi_square_test(data, col1, col2):
    # Create a contingency table
    contingency_table = pd.crosstab(data[col1], data[col2])
    
    # Perform the Chi-Square test
    chi2_statistic, p_value, _, _ = chi2_contingency(contingency_table)
    
    # Define null and alternative hypotheses
    null_hypothesis = f"H0: There is no significant relationship between {col1} and {col2}."
    alternative_hypothesis = f"H1: There is a significant relationship between {col1} and {col2}."
    
    # Print the results
    print(f"\nAnalysis for {col1} and {col2}")
    print("Contingency Table:")
    print(contingency_table)
    print("\nChi-square statistic:", chi2_statistic)
    print("p-value:", p_value)
    
    # Compare p-value with significance level
    alpha = 0.05
    if p_value < alpha:
        print(f"\nReject null hypothesis. {alternative_hypothesis}")
    else:
        print(f"\nFail to reject null hypothesis. {null_hypothesis}")
    
    # Calculate degrees of freedom
    degrees_freedom = (contingency_table.shape[0] - 1) * (contingency_table.shape[1] - 1)
    print(f"\nDegrees of Freedom: {degrees_freedom}")
    
    # Critical value from the chi-square distribution table
    critical_value = chi2.ppf(1 - alpha, degrees_freedom)
    print(f"Critical Value: {critical_value}")
    
    # Compare chi-square statistic with critical value
    if chi2_statistic > critical_value:
        print(f"Reject null hypothesis. {alternative_hypothesis}")
    else:
        print(f"Fail to reject null hypothesis. {null_hypothesis}")
    
    # Print the significance level
    print(f"Significance Level: {alpha}")



# Analyze the relationship between Course and Depression
perform_chi_square_test(data, 'Course', 'Depression')
```
